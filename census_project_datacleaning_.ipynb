```python
import pandas as pd

filepath = "C:/users/lenovo"
filename = "census_2011.xlsx"
filePathFileName = filepath + "/" + filename
outputPathFileName = filepath + "/cleaned_" + filename
outputFileName = "cleaned_" + filename

df = pd.read_excel(filePathFileName, header=0)

#Task 1 : Renaming columns

df.rename(columns={"State name":"State/UT","District name":"District","Male_Literate":"Literate_Male",
"Female_Literate":"Literate_Female","Rural_Households":"Households_Rural","Urban_ Households":"Households_Urban",
"Age_Group_0_29":"Young_and_Adult", "Age_Group_30_49" :"Middle_Aged","Age_Group_50":"Senior_Citizen",
"Age not stated":"Age_Not_Stated"}, inplace = True)
df.rename(columns={"Urban_ Households":"Households_Urban"}, inplace = True)
df['State/UT'] = df['State/UT'].str.title() 

#Task 3 : Rename the State/UT From “Andhra Pradesh” to “Telangana” 
df['State/UT'].replace('Adilabad','Telagana', inplace=True)
df['State/UT'].replace('Nizamabad', 'Telangana', inplace=True)
df['State/UT'].replace('Karimnagar', 'Telangana', inplace=True)
df['State/UT'].replace('Medak', 'Telangana', inplace=True)
df['State/UT'].replace('Hyderabad', 'Telangana', inplace=True)
df['State/UT'].replace('Rangareddy', 'Telangana', inplace=True)
df['State/UT'].replace('Mahbubnagar', 'Telangana', inplace=True)
df['State/UT'].replace('Nalgonda', 'Telangana', inplace=True)
df['State/UT'].replace('Warangal', 'Telangana', inplace=True)
df['State/UT'].replace('Khammam', 'Telangana', inplace=True)
#print(df.duplicated())

def rename_State(state_name) :
    if state_name=='Adilabad' or 'Nizamabad' or 'Karimnagar' or 'Medak' or 'Hyderabad' or 'Rangareddy' or 'Mahbubnagar'  or 'Nalgonda' or 'Warangal' or 'Khammam' :
         return 'Telangana'
    else :
         return "Andhra Pradesh"

#df['State/UT']=df['State/UT'].apply(rename_State)
#print(df["State/UT"].count)
df['State/UT'].replace('Leh(Ladakh)','Ladakh', inplace=True)
df['Population']=df['Population'].fillna(df['Male'] + df['Female'])
df['Male']=df['Male'].fillna(df['Population'] - df['Female'])
df['Female']=df['Female'].fillna(df['Population'] - df['Male'])
df['Literate']=df['Literate'].fillna(df['Literate_Male'] + df['Literate_Female'])
df['Literate_Male']=df['Literate_Male'].fillna(df['Literate'] - df['Literate_Female'])
df['Households']=df['Households'].fillna(df['Households_Rural'] + df['Urban_Households'])
df['Households_Rural']=df['Households_Rural'].fillna(df['Households'] - df['Urban_Households'])
df['Urban_Households']=df['Urban_Households'].fillna(df['Households'] - df['Households_Rural'])
df['SC']=df['SC'].fillna(df['Male_SC'] + df['Female_SC'])
df['Female_SC']=df['Female_SC'].fillna(df['SC'] - df['Male_SC'])
df['Male_SC']=df['Male_SC'].fillna(df['SC'] - df['Female_SC'])
df['ST']=df['ST'].fillna(df['Male_ST'] + df['Female_ST'])
df['Female_ST']=df['Female_ST'].fillna(df['ST'] - df['Male_ST'])
df['Male_ST']=df['Male_ST'].fillna(df['ST'] - df['Female_ST'])
df['Marginal_Workers']=df['Marginal_Workers'].fillna(df['Workers'] - df['Main_Workers'])
#df['Main_workers']=df['Main_workers'].fillna(df['Male_workers'] + df['Female_workers'])
print(df.isna().any()[lambda x: x])
#df.to_excel('cleaned_data.xlsx',index=False,engine='openpyxl')
```
